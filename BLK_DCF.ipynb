```python
import pandas as pd
import yfinance as yf
import numpy as np
import requests
import matplotlib.pyplot as plt
from datetime import datetime 

ticker = 'BLK'
stock = yf.Ticker(ticker)
```

```python
income_statement = stock.financials
balance_sheet = stock.balance_sheet
cash_flow = stock.cashflow
shares_outstanding = stock.info["sharesOutstanding"]

print("Income Statement:\n", income_statement)
print("Balance Sheet:\n", balance_sheet)
print("Cash Flow:\n", cash_flow)
print("Shares Outstanding\n", shares_outstanding)


df_bal = pd.DataFrame(balance_sheet).T
df_inc = pd.DataFrame(income_statement).T
df_cfs = pd.DataFrame(cash_flow).T
df_all = pd.concat([df_bal, df_inc, df_cfs], axis=1).sort_index()
df_all = df_all.loc[:, ~df_all.columns.duplicated()]
```

```python
df = df_all.iloc[-10 :, 1:,].map(lambda x: float(x)/1_000_000 if pd.notna(x) else np.nan)
df.index = pd.to_datetime(df.index)
df.index = df.index.year
df.head(10)
```

```python
revenue = df['Total Revenue'] if 'Total Revenue' in df.columns else df['Revenue']
operating_income = df['Operating Income']
ebit = df['EBIT'] if 'EBIT' in df.columns else operating_income
tax_expense = df['Tax Provision']
capex = df['Capital Expenditure'] * -1  
depreciation = df['Depreciation And Amortization']

# Calculate Free Cash Flow
fcf = operating_income - tax_expense + depreciation + capex
print("Free Cash Flow by year:")
print(fcf.dropna())
```
